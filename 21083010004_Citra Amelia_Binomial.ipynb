```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom
import warnings
warnings.filterwarnings("ignore")
```

```python
# Membangkitkan binomial melalui uniform
unif = np.random.uniform(size=1000)
def inverse(unif, n, p):
    return binom.ppf(unif, n, p) #rumus inverse cdf binomial dist
```

```python
# Hasil dari inverse method binomial 
sns.distplot(inverse(unif, 10, 0.5))
plt.show()

# Hasil dari default numpy
sns.distplot(np.random.binomial(10, 0.5, size=1000))
plt.show()
```

```python
#membandingkan plot
fig = plt.figure(figsize=(20,12))
sns.distplot(inverse(unif, 10, 0.5))
sns.distplot(np.random.binomial(10, 0.5, size=1000))
fig.legend(labels=['inverse method','default numpy'])
plt.show()
```
