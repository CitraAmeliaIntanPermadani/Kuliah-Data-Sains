### Menghapus missing value

```python
#menghapus baris DKI Jakarta sebagai penanganan missing value yang nilainya NaN semua
df_cleaned = kelapa.drop(index=10)
df_cleaned
```

```python
#menghapus baris indeks pertama dari data terakhir karena NaN semua
misval = df_cleaned.dropna(subset=df_cleaned.columns[:-1], how='all')
misval
```

### Mengganti missing value dengan rata-rata kolom

```python
#mengganti nilai NaN pada data kolom 2016 dengan rata-rata data 2016
#menghasilkan nilai rata-rata sebesar 301.006037
a = np.mean(misval['2016'])
print(a)
tabel = misval.fillna({'2016': a})
tabel
```

### Menghapus data yang tidak diperlukan

```python
#Menghapus No. yang tidak diperlukan / tdk relevan
del tabel['No.']
tabel
```

### Menambahkan baris baru dari hasil penjumlahan seluruh data

```python
# tambahkan baris total dari total (sum) setiap variabel data
tabel.loc['Total'] = ['Indonesia', tabel['2014'].sum(), tabel['2015'].sum(), tabel['2016'].sum(), tabel['2017'].sum(),
                   tabel['2018'].sum(), tabel['2019'].sum(), tabel['2020'].sum(), tabel['2021'].sum(),
                   (tabel.iloc[1]['Pertumbuhan 2019 over 2018 (%)'] / tabel.iloc[0]['Pertumbuhan 2019 over 2018 (%)']) - 1]

# cetak DataFrame hasil penggantian isi baris
tabel
```

```python
#menghapus baris dengan Index ke 35 (Indonesia) karena telah diganti dengan total
hapus = tabel.drop(index=35)
hapus
```

```python
#membulatkan kolom Pertumbuhan 2019 over 2018 (%) dengan 2 angka dibelakang koma seperti data awal
hapus['Pertumbuhan 2019 over 2018 (%)'] = hapus['Pertumbuhan 2019 over 2018 (%)'].round(2)
hapus
```

# Detecting Outlier Dengan Boxplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
kelapa_bersih = kelapa_bersih[['2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', 'Pertumbuhan 2019 over 2018 (%)']]
```

```python
plt.figure(figsize=(10,10))
kelapa_bersih.boxplot()
plt.show()
```

```python
# calculate the IQR
Q1 = kelapa_bersih.quantile(0.25)
Q3 = kelapa_bersih.quantile(0.75)
IQR = Q3 - Q1

# calculate the upper and lower bounds
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR

# create a boolean array that indicates which rows are outliers
outlier_mask = ((kelapa_bersih <= lower) | (kelapa_bersih >= upper)).any(axis=1)
print("Detect which row have outlier :")
print(outlier_mask)

# filter the DataFrame to remove outliers
#the negation operator (~) to select the rows that are not outliers.
data_filtered = kelapa_bersih.loc[~outlier_mask]
data_filtered = data_filtered.reset_index(drop=True)
print("Data after remove outlier :")
data_filtered
```

Double Checking

```python
#deteksi dgn IQR value

Q1 = np.percentile(data_filtered,25, interpolation ='midpoint')

Q3 = np.percentile(data_filtered, 75, interpolation = 'midpoint')

IQR = Q3 - Q1

#above upper bound
upper = Q3+1.5*IQR
upper_array1 = np.array(data_filtered>=upper)
print("Upper bound: ", upper, "\n")
print(upper_array1)

#Below Lower bound
lower = Q1-1.5*IQR
lower_array1 = np.array(data_filtered<=lower)
print("lower bound: ", lower, "\n")
print(lower_array1)
```
