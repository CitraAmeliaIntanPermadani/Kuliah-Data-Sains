```python
def scrape_detik(hal):
    global hades
    a = 1
    for page in range(1, hal):
        url = f'https://www.detik.com/search/searchnews?query=pemilu+2024&sortby=time&page={page}'
        ge = req.get(url, hades).text
        sop = bs(ge, 'lxml')
        li = sop.find('div', class_='list media_rows list-berita')
        lin = li.find_all('article')
        for x in lin:
            link = x.find('a')['href']
            date = x.find('a').find('span', class_='date').text.replace('WIB', '').replace('detikNews', '').split(',')[1]
            headline = x.find('a').find('h2').text
            ge_ = req.get(link, hades).text
            sop_ = bs(ge_, 'lxml')
            content = sop_.find_all('div', class_='detail__body-text itp_bodycontent')
            for x in content:
                x = x.find_all('p')
                y = [y.text for y in x]
                content_ = ''.join(y).replace('\n', '').replace('ADVERTISEMENT', '').replace('SCROLL TO RESUME CONTENT', '')
                category = sop_.find('div', class_='detail_tag')
                author = sop_.find('div', class_='detail_author')
                if category is not None:
                    category = category.text.strip()
                else:
                    category = ""
                if author is not None:
                    author = author.text.strip()
                else:
                    author = ""
                print(f'done[{a}] > {headline[0:10]}')
                a += 1
                with open('politikbaru.csv', 'a', encoding='utf-8') as file:
                    wr = csv.writer(file, delimiter=',')
                    wr.writerow([headline, category, date, author, link, content_])
```

```python
import pandas as pd
import re
from bs4 import BeautifulSoup
import requests

class NewsScraper:
    def __init__(self, url):
        self.url = url
        self.data = {
            'Kategori': [],
            'Author': [],
            'Tanggal & Waktu Publish': [],
            'Isi Content': []
        }

    def scrape(self):
        response = requests.get(self.url)
        soup = BeautifulSoup(response.content, 'html.parser')

        links = soup.find_all('a')
        for link in links:
            com = link.get('dtr-evt')
            if com == 'header' and link.get('dtr-sec') == 'first navbar':
                pass
            if com == 'breadcrumb' and link.get('dtr-sec') == 'breadcrumbkanal':
                self.data['Kategori'].append(link.text.strip())

        author = soup.find('div', class_='detail__author')
        self.data['Author'].append(author.text.strip().split(' - ')[0])

        date = soup.find('div', class_='detail__date')
        self.data['Tanggal & Waktu Publish'].append(date.text.strip())

        content = soup.find('div', class_='detail__body-text itp_bodycontent')
        content_text = content.text.replace('\n', '').replace('ADVERTISEMENT', '').replace('SCROLL TO RESUME CONTENT', '')
        hasil_content = re.sub(r'\.\s+', '. ', content_text)
        self.data['Isi Content'].append(hasil_content)

    def to_dataframe(self):
        df = pd.DataFrame(self.data)
        return df

# Gunakan kelas NewsScraper untuk scraping dan membuat dataframe
url = 'https://news.detik.com/berita/d-6760056/dpd-anggarkan-rp-144-m-perbaiki-ruang-anggota-dan-rp-48-m-renovasi-toilet'
scraper = NewsScraper(url)
scraper.scrape()
df = scraper.to_dataframe()

# Cetak dataframe
df
```
